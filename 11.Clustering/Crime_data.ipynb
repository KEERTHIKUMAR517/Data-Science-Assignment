# Data Preparation

```python
X = crime.drop('Unnamed: 0',axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X_scaled = minmax.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(X_scaled)
X_scaled
```

# Using Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(X_scaled,method='complete',))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
```

```python
hc = AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='complete')
hc
```

```python
y_hc = hc.fit_predict(X)
y_hc_df = pd.DataFrame(y_hc)
y_hc_df
```

```python
crime['clusters']  = y_hc_df
crime
```

# Using Kmeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss =  []

for i in range (1,11):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('NO. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X_scaled)
```

```python
y_pred= kmeans.labels_
y_pred
```

```python
crime2 = crime.copy()
```

```python
crime2['clusters'] = y_pred
crime2
```

```python
crime2.sort_values(by = ['clusters'],ascending=False,axis = 0)
```

```python
km = kmeans.cluster_centers_
km
```

```python
crime2
```

```python
c1 = crime2[crime2['clusters'] ==0]
c2 = crime2[crime2['clusters'] ==1]
c3 = crime2[crime2['clusters'] ==2]
c4 = crime2[crime2['clusters'] ==3]

sns.scatterplot(c4.Murder,c4.Assault,color = 'red')
sns.scatterplot(c1.Murder,c1.Assault,color = 'green')
sns.scatterplot(c2.Murder,c2.Assault,color = 'black')
sns.scatterplot(c3.Murder,c3.Assault,color = 'purple')
plt.show()
```

# Using  DBSCAN Clustering

```python
crime3 = crime.copy()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.25,min_samples=2)
db_pred = db.fit_predict(X_scaled)
```

```python
crime3['clusters'] = db_pred
crime3
```

```python
c1 = crime3[crime3['clusters'] ==0]
c2 = crime3[crime3['clusters'] ==1]
c3 = crime3[crime3['clusters'] ==2]
c4 = crime3[crime3['clusters'] ==3]
c5 = crime3[crime3['clusters'] ==-1]

sns.scatterplot(c4.Murder,c4.Assault,color = 'red')
sns.scatterplot(c1.Murder,c1.Assault,color = 'green')
sns.scatterplot(c2.Murder,c2.Assault,color = 'black')
sns.scatterplot(c3.Murder,c3.Assault,color = 'purple')
sns.scatterplot(c5.Murder,c5.Assault,color = 'grey')
plt.show()
```
