# Import Necessary Library

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Collection

```python
movie = pd.read_csv('my_movies.csv')
movie
```

```python
df = movie.drop(['V1','V2','V3','V4','V5'],axis = 1)
df
```

# Model Building

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

# Associates with support: 0.1, confidence:80

```python
freq_movie = apriori(df, min_support=0.1,use_colnames=True)
freq_movie
```

```python
associates = association_rules(freq_movie,min_threshold=0.8,metric = 'lift')
associates
```

```python
x =associates['support']
y = associates['confidence']
z = associates['lift']
```

```python
sns.set_style('darkgrid')
sns.scatterplot(data=associates,x = x,y = y,hue=z,palette='CMRmap')
plt.show()
```

```python
sns.scatterplot(data=associates,x = x,y = z,color = 'blue',hue = y,palette='CMRmap')
plt.show()
```

```python
freq_movie = apriori(df, min_support=0.15,use_colnames=True)
freq_movie
```
